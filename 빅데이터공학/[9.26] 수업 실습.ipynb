```python
# 특정값만 변경
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df

#a 컬럼의 모든 값을 10으로 바꾸자

df['A'] = 10
df
```

```python
## 조건에 따라 값을 변경하는 방법
## B 컬럼값이 5보다 큰 경우에, A를 20으로 바꿔줘
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df.loc[df['B'] > 5, 'A'] = 20
df
```

```python
#인덱스 1의 행, A랑B를 바꿔줘

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df.loc[1,['A','B']] = [50,60]
df
```

```python
# A 컬럼의 값이 50인 행의 B를 100으로 변경해줘 

df.loc[df['A']==50, 'B'] = 100
df
```

```python
# apply를 활용한 값 업데이트
# 'A'컬럼의 값에 5를 곱한 결과를 다시 c컬럼을 만들어서 저장 

df['C'] = df['A'].apply(lambda x: x*5)
df
```
