```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Iris 데이터셋 로드
from sklearn.datasets import load_iris

# Iris 데이터 로드
iris = load_iris()
data = iris.data
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_data['class'] = iris.target

# 데이터 개수 (Count)
count = len(iris_data)
print(f"데이터 개수: {count}")
```

```python
# 데이터 개수 백분율 (Count %)
count_percentage = count / len(iris_data) * 100
print(f"데이터 개수 백분율: {count_percentage:.2f}%")
```

```python
# 최솟값 (Min)
min_values = iris_data.min()
print("최솟값:")
print(min_values)

# 최댓값 (Max)
max_values = iris_data.max()
print("최댓값:")
print(max_values)

# 평균 (Mean)
mean_values = iris_data.mean()
print("평균:")
print(mean_values)

# 중앙값 (Median)
median_values = iris_data.median()
print("중앙값:")
print(median_values)

# 최빈값 (Mode)
mode_values = iris_data.mode().iloc[0]
print("최빈값:")
print(mode_values)

# 범위 (Range)
range_values = iris_data.max() - iris_data.min()
print("범위:")
print(range_values)

# 사분위수 (Quantiles)
quantiles = iris_data.quantile([0.25, 0.5, 0.75])
print("사분위수:")
print(quantiles)

# 분산 (Variance)
variance_values = iris_data.var()
print("분산:")
print(variance_values)

# 표준 편차 (Standard Deviation)
std_deviation_values = iris_data.std()
print("표준 편차:")
print(std_deviation_values)
```

```python
# 변동계수 : 데이터와의 상대적인 변동성을 나타내는 통계적 측정 지표
# 주로 표준편차와 평균사이의 관계를 사용해 계상함
# 주요 사용처: 다른 데이터 집합과의 비교, 서로 다름 데이터 집합의 상대적인 변동성을 비교할때 유용, 
# 변동계수가 작은 데이터 집합은 상대적으로 일정한 경향을 보이며, 변동계수가 큰 데이터 집합은 상대적으로 불안정하고 
# 큰 변동을 나타냄
# 단위가 다른 데이터 비교 : 서로 다른 측정 단위를 가지는 데이터를 비교할 때 사용,
# 변동계수는 상대적인 변동성을 단위에 독립적으로 나타내므로, 단위가 다른 데이터간에도 비교가 가능함
# 두개의 주식 포트폴리오가 있을때, 포트폴리오 A의 변동계수가 B의 변동계수보다 작다면, A는 비교적 안정적이고 덜 위험함
# 변동계수는 데이터의 상대적인 변동성을 빠르게 이해하고 비교하는데 도움을 주며 특히 금융, 경제, 공학 및 자연과학분야에 널리 사용됨
```

```python
# 변동 계수 (Coefficient of Variation)
coef_of_variation = (std_deviation_values / mean_values) * 100
print("변동 계수:")
print(coef_of_variation)
```

```python
# 왜도(Skewness)와 첨도(Kurtosis)는 확률 분포나 데이터 분포의 모양을 설명하는 중요한 통계적 지표입니다. 
# 이 두 지표는 데이터의 분포가 얼마나 비대칭적이고 뾰족한지를 나타냅니다.

# 왜도 (Skewness):

# 왜도는 데이터 분포의 비대칭성을 측정하는 지표입니다.
# 왜도 값이 0보다 작으면 데이터 분포가 왼쪽으로 비대칭 (음의 왜도), 0보다 크면 오른쪽으로 비대칭 (양의 왜도), 0에 가까우면 대칭적인 분포입니다.
# 왜도는 분포의 꼬리(tail)가 어느 쪽으로 길게 늘어져 있는지를 나타냅니다. 음의 왜도의 경우 왼쪽 꼬리가 길게 늘어져 있고, 양의 왜도의 경우 오른쪽 꼬리가 길게 늘어져 있습니다.
# 첨도 (Kurtosis):

# 첨도는 데이터 분포의 뾰족함을 나타내는 지표입니다.
# 첨도 값이 3에 가까우면 정규 분포와 비슷한 뾰족한 분포를 나타냅니다. 이를 정규 분포(standard normal distribution)라고 합니다.
# 첨도 값이 3보다 크면 분포가 정규 분포보다 뾰족한 분포를 가지며, 이러한 분포를 "양수 첨도" 또는 "뾰족한 분포"라고 합니다. 양수 첨도의 경우 꼬리 부분이 뾰족하게 뾰족한 분포를 나타냅니다.
# 첨도 값이 3보다 작으면 분포가 정규 분포보다 더 완만한 분포를 가지며, 이러한 분포를 "음수 첨도" 또는 "완만한 분포"라고 합니다.
# 왜도와 첨도는 데이터 분포의 모양을 묘사하고 분포의 특성을 이해하는 데 도움을 줍니다. 이러한 지표는 통계 분석 및 데이터 시각화에서 중요한 역할을 하며, 데이터 분포가 어떤 형태를 띄는지 파악하는 데 도움이 됩니다.
```

```python
# 왜도 (Skewness)
skewness = iris_data.skew()
print("왜도:")
print(skewness)

# 첨도 (Kurtosis)
kurtosis = iris_data.kurtosis()
print("첨도:")
print(kurtosis)
```

```python
### 카이제곰 검정 (Chi-squared test)
# 클래스(종)와 꽃받침 길이 간의 관계를 검정합니다. 검정 결과와 p-value를 출력합니다.

from scipy.stats import chi2_contingency

# 클래스 (Species)와 꽃받침 길이 (Sepal Length) 간의 관계를 검정합니다.
contingency_table = pd.crosstab(iris_data["class"], pd.qcut(iris_data["sepal length (cm)"], 3))
chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-squared test 결과:")
print(f"Chi-squared: {chi2}")
print(f"P-value: {p}")
```

### 데이터 품질검사, 변수선택등

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler, RobustScaler
from sklearn.datasets import load_iris

# Iris 데이터 로드
iris = load_iris()
data = iris.data
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

#### 파생변수

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# 파생 변수 생성 예시:

# 1. 꽃받침(sepal)의 넓이: 꽃받침 길이 x 꽃받침 너비
iris_df['sepal_area'] = iris_df['sepal length (cm)'] * iris_df['sepal width (cm)']

# 2. 꽃잎(petal)의 넓이: 꽃잎 길이 x 꽃잎 너비
iris_df['petal_area'] = iris_df['petal length (cm)'] * iris_df['petal width (cm)']

# 3. 꽃받침과 꽃잎의 길이 비율
iris_df['sepal_to_petal_length_ratio'] = iris_df['sepal length (cm)'] / iris_df['petal length (cm)']

# 4. 꽃받침과 꽃잎의 너비 비율
iris_df['sepal_to_petal_width_ratio'] = iris_df['sepal width (cm)'] / iris_df['petal width (cm)']

iris_df.head()
```

#### 스케일링

데이터 스케일링은 변수들 간의 척도 차이를 줄여, 알고리즘의 성능을 향상시키는 중요한 전처리 단계

```python


# Min-Max 스케일링 (정규화) 각 특성의 값을 0과 1 사이의 범위로 조정합니다.
min_max_scaler = MinMaxScaler()
data_minmax = min_max_scaler.fit_transform(data)
print("Min-Max Scaling (Normalization):")
print(data_minmax[:5])  # 첫 5개 행만 출력

# MaxAbs 스케일링
#데이터를 절대값이 0~1 범위에 있도록 변환합니다. 다시 말해, 각 특성에 대해 데이터는 해당 특성의 최대 절대값으로 나누어집니다. 
#이는 이미 중심이 0에 맞춰진 데이터에 유용합니다. 특히, 희소 데이터(sparse data)에서도 그 중심을 유지하려 할 때 사용
max_abs_scaler = MaxAbsScaler()
data_maxabs = max_abs_scaler.fit_transform(data)
print("MaxAbs Scaling:")
print(data_maxabs[:5])  # 첫 5개 행만 출력

# Z-score 스케일링 (표준화) 각 특성의 평균을 0, 표준편차를 1로 조정하여 데이터를 표준화합니다.
standard_scaler = StandardScaler()
data_standard = standard_scaler.fit_transform(data)
print("\nZ-score Scaling (Standardization):")
print(data_standard[:5])  # 첫 5개 행만 출력

# Robust 스케일링 (IQR을 사용한 스케일링) 중앙값과 IQR(Interquartile Range)을 사용하여 아웃라이어의 영향을 최소화하면서 스케일링합니다.
robust_scaler = RobustScaler()
data_robust = robust_scaler.fit_transform(data)
print("\nRobust Scaling:")
print(data_robust[:5])  # 첫 5개 행만 출력
```

```python
### Bining
##변수의 구간화(binning)는 연속적인 변수를 여러 구간으로 나누는 과정입니다. 
## 구간화는 특정 변수의 분포나 관계를 보다 쉽게 이해할 수 있게 도와주거나, 범주형 변수로 변환할 때 사용됩니다.

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# 'sepal length (cm)' 특성을 4개의 구간으로 나눔
bins = [4, 5, 6, 7, 8]
labels = ['4-5', '5-6', '6-7', '7-8']
iris_df['sepal_length_bin'] = pd.cut(iris_df['sepal length (cm)'], bins=bins, labels=labels)

print(iris_df[['sepal length (cm)', 'sepal_length_bin']].head(10))
```

#### 변수변환

변수 변환은 주어진 데이터의 형태나 분포를 변경하기 위한 방법으로, 선형성 개선, 이상치의 영향 최소화, 데이터의 분포 조정 등의 목적으로 사용됩니다. 대표적인 변환 방법에는 로그 변환, 제곱근 변환, 역수 변환 등이 있습니다.

```python
# 'sepal length (cm)' 특성에 대한 로그 변환
iris_df['log_sepal_length'] = np.log(iris_df['sepal length (cm)'])

# 원래의 'sepal length'와 로그 변환된 'log_sepal_length'의 분포 시각화
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

iris_df['sepal length (cm)'].hist(ax=ax[0], bins=30)
ax[0].set_title('Original Sepal Length')
ax[0].set_xlabel('Sepal Length (cm)')
ax[0].set_ylabel('Frequency')

iris_df['log_sepal_length'].hist(ax=ax[1], bins=30)
ax[1].set_title('Log-transformed Sepal Length')
ax[1].set_xlabel('Log Sepal Length')
ax[1].set_ylabel('Frequency')

plt.tight_layout
```

```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA

# 데이터 불러오기
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# PCA를 사용하여 데이터를 2차원으로 축소
pca = PCA(n_components=2)
X_r = pca.fit_transform(X)

# 결과를 시각화
plt.figure(figsize=(8, 6))

colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.show()
```
